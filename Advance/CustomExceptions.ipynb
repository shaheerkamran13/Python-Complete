# Custom Exeption ( Raising and throwing an exception)

```python

class Error(Exception):
    pass

class dobException(Error):
    pass

```

```python
year = int(input("Input the birth year"))
age = 2026-year


try:
    if age<=30 and age>=20:
        print("The age is valid so you can apply for the exam")

    else:
        raise dobException

except dobException:
    print("Sorry your age should be greater then 20 and less then 30")
```
