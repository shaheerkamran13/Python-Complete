# Iterators

- `Iterators` are advanced Python concepts that allow for efficient looping and memory management. 

- Iterators provide a way to access elements of a collection sequentially without exposing underlying structure


```python
my_list =   [1,2,3,4,5,6]

for i in my_list:
    print(i)

type(my_list)
```

```python
## Iterator

iterator = iter(my_list)
print(type(iterator))
```

```python
iterator
```

```python
my_list_abc =   "Hello World"
iterator1 = iter(my_list_abc)

try:
    print(next(iterator1))

except StopIteration:
    print("Hey Stop the Iteration, no elements in the iterator")

print(next(iterator1))
```
