# Pandas

- Data Manipulation Library
- Widely used in data analysis and data cleaning

- Provides Series and DataFrame.

- Series: A one-dimensional array  like object.

- DataFrame: A two-dimensional, size-mutable and heterogenous tabular data structure with labeled axes (rows and columns)

```python
# Series
# A panadas series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table

import pandas as pd

data = [1,2,3,4,5]
series = pd.Series(data)
print('Series: \n', series)
print(type(series))
```

```python
# Create a series from dictionary
data = {'a': 1, 'b': 2, 'c': 3}

series_dict =  pd.Series(data)
print(series_dict)
```

```python
data = [10,20,30]

index = ['a','b','c']
pd.Series(data,index=index)
```

```python
# DataFrame
# Create a DataFrame from a dictionary of list

data = {
    'Name': ["Bob", 'Tom', 'John'],
    'Age': [25,30,45],
    'City': ['Multan','Islamabad', 'Lahore']
}


df = pd.DataFrame(data)
print(df)

print(type(df))
```

```python
import numpy as np

np.array(df)
```

```python
# Create a dataframe from a list of dictionaries

data = [
    {'Name': "Bob", 'Age':23, 'City': 'Islamabad'},
    {'Name': "John", 'Age':20, 'City': 'Islamabad'},
    {'Name': "Bappy", 'Age':32, 'City': 'Islamabad'},
    {'Name': "Jack", 'Age':19, 'City': 'Islamabad'},

]

df = pd.DataFrame(data)
print(df)

print(type(df))
```
