```python
# Filling missing values with the mean of column

df['album_artist_fillNA'] = df['album_artist'].fillna(df['album_artist'].mean())
```

```python
df.dtypes
```

```python
# Renaming columns
df = df.rename(columns={'track_name':'Song_Names'})

df.head()
```

```python
# Change Datatypes

df['Value_new'] = df['track_time_millis'].fillna(df['track_time_millis']).mean().astype(float)
df.head()
```

```python
df['New___Value'] = df['collection_price'].apply(lambda x:x*2)

df.head()
```

```python
# Merging and joining dataframes

df1 = pd.DataFrame({'Key':['A', 'B', 'C'], 'Value1': [1,2,3]})

df2 = pd.DataFrame({'Key':['A', 'B', 'D'], 'Value2': [4,5,6]})
```

```python
df1
```

```python
df2
```

```python
# Merge DataFrame on the 'Key Columns'ArithmeticError
pd.merge(df1, df2, on='Key', how='inner')
```

```python
pd.merge(df1, df2, on='Key', how='outer')
```

```python
pd.merge(df1, df2, on='Key', how='left')
```

```python
pd.merge(df1, df2, on='Key', how='right')
```
