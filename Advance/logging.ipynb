# Python Logging

- Track errors, events, operational information.

# Log Levels

Python got different logging modules having several log levels, indicating severity of events.

The levels are:

- DEBUG: Detailed Information, of interest when diagnosing problems.
- INFO: Confirmation of things working as expected.
- WARNING: Indication of something unexpected happened or indicative of some problem in near future. Software still works as expected.
- ERROR: Due to a serious problem, the software has not been able to perform some function
- CRITICAL: A very serious error, indicating the program itself may be unable to continue running

```python
import logging

logging.basicConfig(

filename='shaheer.log',
filemode='w',
level = logging.DEBUG ,
format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
datefmt='%Y-%m-%d %H:%M:%S'
)


# Log messages with different severity levels

logging.debug("This is a debug message")
logging.info('This is a info message')
logging.warning("This is a warning message")
logging.error('This is a error message')
logging.critical('THis is a critical message')
logging.critical('THis is a critical message.....')
```
