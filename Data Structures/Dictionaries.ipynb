# Dictionaries
- Key-value pairs
- Key must be unique and immutable (i.e strings, number and tuples)
- Value can be of any type


```python
# Creating Dictionaries

empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict = dict()
print(type(empty_dict))
```

```python
student = {"name":"Shaheer", "age":23,"grade":24}
print(student)
print(type(student))
```

```python
# Error

# Single key is always used
student = {"name":"Shaheer", "age":23,"grade":24, "name":"Kamran"}
print(student)
```

```python
# Accessing dictionary Elements

student = {"name":"Shaheer", "age":23,"grade":"A"}
print(student["age"])
print(student["name"])

# Accsessing using get() method
print(student.get('age'))
print(student.get('last_name'))
print(student.get('last_name', "Key Not Found"))



```

```python
# Modifying Dictionary elements
# Dictionary are mutable, elements can be added, updated or deleted.
# Keys should be unique all the time

print(student)

student["age"] = 45   # Updated

print(student["age"])

student['address'] = "House-7D New Shamsabad, Multan" # Added a new key value
print(student)


```

```python
del student['grade'] # Delete key value pair

print(student)
```

```python
# Dictionary Methods

keys = student.keys() # get all the keys
print(keys)
values = student.values() # get all the values
print(values)


items = student.items() # get all the items
print(items)
```

```python
# Shallow Copy

student_copy = student 
print(student)
print(student_copy)
```

```python
student["name"] = "Shaheer-Kamran-H" 
print(student)
print(student_copy)  # Assigning of value to one key of a dictionary is happening same to another key of other dictionary
```

```python
# Shallow Copy

student_copy1 = student.copy()
print(student_copy1)
print(student)
```

```python
# Iterating over disctionaries
# You can us e loops to iterate over dictionaries. i.e key, value or items


# iterating over keys


for key in student.keys():
    print(key)


# iterating over values


for value in student.values():
    print(value)    

# iterating over items


for item in student.items():
    print(item)
```

```python
# Key-Value pair iterating

for key,value in student.items():
    print(f"{key}: {value}")
```

```python
# Nested Dictionaries

students = {
    "student1": {"name":"Shaheer", "age": 23},
    "student2": {"name":"Kamran", "age": 52},
}

print(students)
```

```python
# Accessing Nested Dictionaries

print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
# Iterating over nested dictionaries

for student_id, student_info in students.items():
    print(f"{student_id}: {student_info}")

    for key, value in student_info.items():
        print(f"{key}: {value}")



```

```python
cubes = {x:x**3 for x in range(10)}
print(cubes)
```

```python
# condition dictionary comprehension
evens = {x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
# Practical Examples

# Use a dicionary to count the frequence of elemetns in list

numbers = [1,2,2,3,3,3,4,4,4,4]

frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] +=1
    else: 
        frequency[number] = 1
print(frequency)
```

```python
# Merge 2 dictionaries into one

dict1 = {"a":1,"b":2}
dict2 = {"b":3,"c":4}

merged_dict = {**dict1, **dict2}
print(merged_dict)
```
