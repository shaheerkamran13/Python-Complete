# Tuples

- Ordered collections of items
- Immutable
- Imuutability differentiates them from lists

```python
# Creating a tuple

empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst = list()
print(type(lst))
print(lst)
tpl = tuple()
print(tpl)
print(type(tpl))
```

```python
numbers = tuple([1,2,3,4,5,6])
numbers

type(numbers)
```

```python
eg = list(numbers)
print(type(eg))
```

```python
mixed_tuple = (1, "Hi shaheer", 3,14, True)
print(mixed_tuple)
```

```python
# Accessing tuple elements
numbers = tuple([1,2,3,4,5,6])
numbers
```

```python
numbers[4]
```

```python
print(numbers[2])
print(numbers[-1])
```

```python
print(numbers[0:4])
print(numbers[::-1])
```

```python
# Tuple Operations 
concatenation_tuple = numbers + mixed_tuple 
print(concatenation_tuple)
```

```python
# Tuple Operations 
numbers * 3
```

```python
# Immutable nature of tuples
# tuples are immutable meaning their elements cannot be changed once assigned

# --- Important --- /

# List are mutable

lst = [1,2,3,4,5]

print(type(lst))

lst[3] = "Shaheer"
print(lst)
```

```python
numbers[1] =  "Kamran" # New elements cannot be assigned to a tuple
```

```python
# Tuple Methods
print(numbers)
print(numbers.count(1))
print(numbers.index(3))
```

```python
# Packing an unpacking tuple

packed_tuple = 1, "hi", 3.14
print(packed_tuple)
```

```python
# Unpacking a tuple

a,b,c = packed_tuple

print(a)
print(b)
print(c)
```

```python
# Unpacking with *
numbers = (1,2,3,4,5,999,8,2,1,6,6)

first, *middle, last = numbers

print(first)
print(middle)
print(last)

```

```python
# Nested Tuple
# Nested List
lst = [[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]

# 
lst[2][0:3]
```

```python
lst = [[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
print(type(lst))
```

```python
nested_tuple = ((1,2,3),('a','b','c'),(True, False))
print(nested_tuple)


# access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][1])
```

```python
# Iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end = " ")
    print()
```
