# File Operations - Read and Write Files

- .strip() method removes the newline character

```python
# Read a whole file

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# read a file line by line

with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

```python
# Write in a file (Overwritting)

with open('example.txt', 'w') as file:
    file.write('Hello World\n')
    file.write('This is a new line')
```

```python
# write a file without overwritting

# use 'a' as an append method to add something new without overwritting


with open('example.txt', 'a') as file:
    file.write('\n---Hey, this is an appended text---')
```

```python
# writing a list of lines to a file

lines = ['First Line\n', 'Second Line\n', 'Third Line\n']

with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
# Binary Files
# In binary files we updates bytes

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'

with open('example.bin','wb') as file:
    file.write(data)
```

```python
# Reading the binary file

with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
# Read the content from a source text file and write to a destination file

# copying to a text file

with open('source.txt', 'r') as file:
    content = file.read()


with open('destination.txt', 'w') as destinationFile:
    destinationFile.write(content)
```

```python
# Read a text file and count the number of lines and characters


def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)

        return line_count, word_count, char_count
    
file_path = 'example.txt'
lines, words, characters = count_text_file(file_path)
print(f"Lines: {lines}, Words: {words}, Characters: {characters}")
```

The w+ mode in python is used to open a file for both reading and writing. If the file does not exist, it will be created. If the file exits, its content is truncated (i.e the file is overwritten)

- Look for .seek() method

```python
# Writing and then reading a file

with open('example.text', 'w+') as file:
     file.write('Hello World\n')
     file.write('This is a new line222\n')

     # move the file cursor to the beginning

     file.seek(0)

     # Read the content of the file
     content = file.read()
     print(content)
```
