# Exception Handling
- Usage of try, except blocks

- Exceptions: The entity which disrupts the normal flow of the program, caused by error.


- Error Examples:
    1. FileNotFoundError
    2. ValueError: Invalid Value
    3. TypeError: Invalid Type



```python
# Exception using try, except block.

try:
    a = b

except:
    print('The variable has not been assigned')
```

```python
try:
    a = b

except NameError as ex:
    print(ex)
```

```python
try:
    result = 1/0
    a=b
except ZeroDivisionError as ze:
    print(ze)
    print("Please enter the denominator greater then 0")
```

```python
try:
    result = 1/2
    a=b
except ZeroDivisionError as ze:
    print(ze)
    print("Please enter the denominator greater then 0")

except Exception as ex1: # Main class to handle all errors
    print(ex1)
    print('Main exception got caught here')
```

```python
try:
    num = int(input("Please enter a number: "))
    result = 10/num

except ValueError:
    print('This is not a valid number')

except ZeroDivisionError:
    print('Enter a number greater then 0')

except Exception as ex:
    print(ex)
```

```python
# try, except, else block

try:
    num = int(input('Please enter a number: '))
    result = 10/num
except ValueError:
    print('The value entered is not a valid number.')

except ZeroDivisionError:
    print('Please enter a value which is greater then 0')

except Exception as ex:
    print(ex)

else:
    print(f'The result of the division is: {result}')
```

```python
# try, else, except, finally


try:
    num = int(input('Please enter a number: '))
    result = 10/num
except ValueError:
    print('The value entered is not a valid number.')

except ZeroDivisionError:
    print('Please enter a value which is greater then 0')

except Exception as ex:
    print(ex)

else:
    print(f'The result of the division is: {result}')
finally:
    print("The exception is complete")
```
