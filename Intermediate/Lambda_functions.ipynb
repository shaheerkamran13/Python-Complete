# Lambda Functions in Python

- Anonymous functions( functions without name)
- Used with lambda keyword
- Can have number of arguments but only one expression.
- Commonly used for short operations 

```python
# Syntax

lambda arguments: expression
```

```python
addition = lambda a,b: a+b

type(addition)
print(addition(5,6))
```

```python
def even(num):
    if num%2 == 0:
        print("Even")


even(2)

even1 = lambda num: num%2==0
even1(2)
```

```python
def addition(x,y,z):
    return x+y+z
addition(23,5,2)
```

```python
addition1 = lambda x1,y1,z1: x1+y1+z1

print(addition1(23,5,2))
```

```python
# map () - It applies a function to all items in a list
numbers = [1,2,3,4,5]

def square(number):
    return number**2

square(2)
```

```python
list(map(lambda x:x**2, numbers))
```
