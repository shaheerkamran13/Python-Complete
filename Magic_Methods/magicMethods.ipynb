# Magic Methods
- Know as `dunder` methods (Double underscore methods). 

- Start and ends with double underscore

- These methods enable you to define the behavior of objects for built-in operations, such as operations, comparisons and more.

# Examples

`__init__` : Initialize a new instance of a class

`__str__` : Returns a string representation of an object.

`__repr__` : Returns an official string representation of an object.

`__len__` : Returns the length of an object.

`__getitem__` : Gets an item from a container.

`__setitem__` : Set an item in a container.

```python
class Person:
    pass

person = Person()
dir(person)
```

```python
print(person)
```

```python
# Basic Magic methods

class Person:
    def __init__(self,name,age):
            self.name = name
            self.age = age
    
    def __str__(self):
          return f"{self.name} is {self.age} years old"
    
    def __repr__(self):
          return f"Person(name = {self.name}, age = {self.age}) "
        
person = Person('Shaheer', 23)

dir(person)
```
