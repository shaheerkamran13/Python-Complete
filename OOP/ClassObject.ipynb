```python
# Instance variables and methods
# __init__ is called constructor
class Dog:
    # constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # create the objects

dog1 = Dog("Pluto",23)
print(dog1)
print(dog1.name)
print(dog1.age)
```

```python
# Instances

dog2 = Dog('Lucy',5)
print(dog2)
print(dog2.name)
print(dog2.age)
```

```python
# Define a class with instance methods

class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says woof")

dog1 = Dog('Buddy',5)
dog1.bark()

dog2 = Dog('Lucy',3)
dog2.bark()
```

```python
# Modeling a bank account

class BankAccount:

    def __init__(self, name, balance=0): # Instance variables
        self.name = name
        self.balance = balance
    
    def deposit(self, amount):
        self.balance = self.balance + amount
        print(f"The balance has been increased and new balance is: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print(f'Insufficient Balance: {self.balance}')
        else:
            self.balance = self.balance - amount
            print(f"This amount is withdrawn. New balance is {self.balance}")
    
    def get_balance(self):
        return self.balance

owner1 = BankAccount('Shaheer', 11000)
print(owner1.balance)


owner1.deposit(500)
owner1.withdraw(120000)
owner1.withdraw(6000)
owner1.get_balance()
```
