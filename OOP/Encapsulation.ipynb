# Access Variables

-  `public`
-  `protected` - Cannot be used outside the class but can be used in a derived class
-  `private` - Cannot be accessed outside the class
-  Access Modifiers

```python
class Person:
    def __init__(self, name, age):
        self.name = name  # public variables
        self.age = age
    

def get_name(person):
    return person.name


person = Person('Shaheer', 23)
print(person.name)

get_name(person)

```

```python
dir(person)
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name  # private variable
        self.__age = age    # private variable
        self.gender = gender
    

def get_name(person):
    return person.name


person = Person('Shaheer', 23, 'Male')
print(person.gender)

dir(person)
```

```python
class Person1:
    def __init__(self, name, age, gender):
        self.__name = name  # private variable
        self.__age = age    # private variable
        self.gender = gender
    

def get_name1(person):
    return person.__name


person1 = Person1('Shaheer', 23, 'Male')
# get_name1(person1)

dir(person1)
```

```python
class Person2:
    def __init__(self, name, age, gender):
        self._name = name  # protected variable
        self._age = age    # protected variable
        self.gender = gender

class Employee(Person2):
    def __init__(self, name, age, gender):
        super().__init__(name,age,gender)



employee = Employee('Shaheer',23, "Male")
print(employee._name)
```

```python
# Encapsulation with getter and setter

class Person3:
    def __init__(self, name, age):
        self.__name = name # Private access modifier or variable
        self.__age = age # Private access variable

    def get_name(self):
        return self.__name
    
    def set_name(self, name):
        self.__name = name

    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative")

person3 = Person3('Shaheer', 23)


# Accessing and modidying private variables thorugh getter and setters

print(person3.get_name())
print(person3.get_age())

person3.set_name('Kamran')
person3.set_age(35)

print(person3.get_name())
print(person3.get_age())
```
