```python
# While Loop
# Runs untill the condition is true

count = 0
while count<5:
    print(count)
    count+=2 
```

```python
# Loop Control Statements

# break: The break exits the loop



# break statement

for i in range(10):
    if (i == 6):
        break
    print(i)
```

```python
# continue: skips the current iteration and moves onto next

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
# pass: The pass operation is a null operation, it does nothing.

for i in range(0,5):
    if i==3:
        print("The number is skipped '3' ")
        pass
    print(i)
```

```python
# Nested Loop

for i in range(3):
    print("Outside Loop: ",i)
    for j in range(5):
        print("Inside Loop: ", j)
    
```

```python
# Example - Calculate the sum of first n natural numbers using while and for loop


# While Loop

num = int(input("Enter the numebr: "))
counter = 0
sum = 0 


while counter<=num:
    sum = sum+counter
    counter +=1

print('The total sum: ', sum)
```

```python
# For Loop

num = int(input('Enter your number: '))

sum = 0
for i in range(0,num+1):
    sum = sum+i

print('The total sum is: ', sum)
```

```python
# Example - Show prime numbers between 1 and 100

for i in range(1,101):
    if i>1:
        for j in range(2, i):
            if i%j == 0:
                break
        else:
            print(i)
```
